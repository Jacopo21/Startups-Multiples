```python
import os
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By

# Set up the web driver (Chrome in this case)
driver = webdriver.Chrome()

# The page to scrape
url = "https://pages.stern.nyu.edu/~adamodar/New_Home_Page/dataarchived.html"
driver.get(url)

# Create the download directory if it doesn't exist
download_dir = "/Users/jacopobinati/Desktop/damodaran"
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Function to download a file given the URL and save path
def download_file(file_url, save_dir):
    file_name = file_url.split("/")[-1]  # Get the file name from URL
    save_path = os.path.join(save_dir, file_name)
    
    # Download and save the file
    response = requests.get(file_url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {file_name}")
    else:
        print(f"Failed to download: {file_name}")

# XPath to get all file links (skip the last row, adjust based on actual HTML structure)
xpaths = [
    '//a[contains(@href, "archives/pedata")]',  # Modify as needed for each row pattern
]

# Collect and download all files except the last row
for xpath in xpaths:
    elements = driver.find_elements(By.XPATH, xpath)
    
    for elem in elements:
        file_url = elem.get_attribute("href")
        download_file(file_url, download_dir)

# Move to the next cell (European data) and download
next_cell_url_europe = "https://pages.stern.nyu.edu/~adamodar/pc/archives/peEurope11.xls"
download_file(next_cell_url_europe, download_dir)

# Move to the next cell (Japan data) and download
next_cell_url_japan = "https://pages.stern.nyu.edu/~adamodar/pc/archives/peJapan11.xls"
download_file(next_cell_url_japan, download_dir)

# Move to the next cell (Emerging Markets data) and download
next_cell_url_emerging = "https://pages.stern.nyu.edu/~adamodar/pc/archives/peemerg11.xls"
download_file(next_cell_url_emerging, download_dir)

# Close the web driver when done
driver.quit()
```
